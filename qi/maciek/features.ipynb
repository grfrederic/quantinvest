```python
funds = pd.read_csv('../../materials/Quant_Invest_Fundusze.csv', sep=';', index_col='Daty', parse_dates=True)
funds.head()
```

```python
data = pd.DataFrame()
for c in funds:
    data['ln_'+c] = np.log(funds[c])
    
data['ln_wig20'] = np.log(wig20)
data['ln_nyse']  = np.log(nyse)

for c in rates:
    data['ln_rates_'+c] = np.log(1+rates[c])

for c in unemp:
    data['unemp_'+c] = unemp[c]

for c in gdp_pct_growth:
    data['ln_gdp_growth_'+c] = np.log(1+gdp_pct_growth[c]/100)
    
data.head()
```

```python
def fourier(x, window_len, n_windows=1):
    '''
    For each point in x, calculate dot product 
    of window_len*n*windows previous(including self) values
    (if neccessary pad with zeroes) and complex exponent
    of period window len. Return absolute values.   
    '''
    # values to convolve with
    v = np.array([complex(np.cos(2*np.pi*k/window_len),
                          np.sin(2*np.pi*k/window_len)) for k in range(window_len)]*n_windows)
    x_ = [0]*(window_len-1) + list(x)
    return np.abs(np.convolve(x, v)[:-window_len*n_windows+1])

fourier(np.sin(2*np.pi*2*np.arange(10)), 5)
```

```python
# calculate periods for fouriering
n_days_per_month = int(len(funds.index)/19/12)
n_days_per_week  = int(len(funds.index)/19/12/4)
n_days_per_month, n_days_per_week
```

```python
transformed = fourier(np.log(funds.AP).diff().dropna(), n_days_per_week)
plt.plot(transformed)
plt.show()
plt.hist(transformed, bins=50)
plt.show()
plt.plot(np.diff(transformed))
plt.show()
```

```python
for c in ['ln_AP', 'ln_OP', 'ln_ARW', 'ln_ORW',
          'ln_ARR', 'ln_ORR', 'ln_G', 'ln_wig20', 'ln_nyse']:
    ticker = c[3:]
    s = data[c]
    ln_growth = s.dropna().diff().dropna()
    fourier_w = fourier(ln_growth, n_days_per_week )
    fourier_m = fourier(ln_growth, n_days_per_month)
    data['ln_growth_'+ticker] = ln_growth
    data.loc[ln_growth.index, 'ln_growth_fw_'+ticker] = fourier_w
    data.loc[ln_growth.index, 'ln_growth_fm_'+ticker] = fourier_m
    del data[c]
```
